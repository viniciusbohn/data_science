## Introdução

### O que é a COVID-19?

A COVID-19 é uma doença respiratória causada pelo coronavírus (SARS-CoV-2). Ela é considerada potencialmente grave, além de possuir elevada transmissibilidade.

Os sintomas mais comuns da doença são a febre, cansaço e tosse seca. Mas em diversos casos registrados, dentre as muitas variações do vírus, foram constatados sintomas como a trombose venosa profunda, AVC (acidente vascular cerebral) e infarto. Isso se deve ao fato da COVID-19 ser considerada uma doença sistêmica. Nos casos mais graves, seu impacto na região pulmonar gera a baixa oxigenação do sangue que, por sua vez, acaba por afetar diversas partes do nosso corpo. 

### Transmissibilidade

Uma das principais características da COVID-19 e que fez, e faz, dela tão temida é sua altíssima transmissibilidade. As pessoas infectadas podem transmitir o vírus por meio da tosse, da fala ou mesmo da respiração, quando expelem gotículas maiores que eventualmente se assentam ou aerossóis menores que podem flutuar no ar. Mesmo que a taxa de mortalidade da COVID-19 não seja muito alta, quando comparamos à outras doenças, sua transmissibilidade faz com que tenhamos muitos casos e, mesmo com um percentual baixo de mortalidade, tenhamos muitas mortes em números brutos.

### Vacinação

Vários laboratórios desenvolveram, em tempo recorde, vacinas de combate à COVID-19. Estas vacinas foram divididas em até três doses e disponibilizadas ao redor do mundo, enfrentando uma série de problemas de produção e distribuição. Há uma série de pessoas que se demonstram resistentes à vacina, por diversos motivos.

### Tratamento dos Dados

Note que a variável `date` não possui o formato de data. Vamos transformá-la no formato `datetime`:

```
df.date = pd.to_datetime(df.date, format = "%Y-%m-%d")

# verificando
df.info()
```

Para facilitar a análise dos dados realmente relevantes, vamos primeiramente eliminar algumas colunas que não são importantes ao estudo:

```
df = df.drop(columns=['weekly_icu_admissions','weekly_icu_admissions_per_million','weekly_hosp_admissions_per_million', 'reproduction_rate', 'aged_65_older', 'aged_70_older', 'extreme_poverty',
                      'weekly_hosp_admissions','excess_mortality','excess_mortality_cumulative_absolute','excess_mortality_cumulative', 'positive_rate', 'gdp_per_capita', 'cardiovasc_death_rate',
                      'total_boosters_per_hundred','total_boosters','excess_mortality_cumulative_per_million','icu_patients_per_million', 'tests_per_case', 'stringency_index',
                      'icu_patients','hosp_patients_per_million','hosp_patients','human_development_index', 'life_expectancy', 'hospital_beds_per_thousand','handwashing_facilities', 'male_smokers', 'female_smokers', 'diabetes_prevalence', 'cardiovasc_death_rate'])
```

Por se tratar de uma das variáveis mais relevantes para o objetivo deste estudo, vamos analisar as entradas únicas para a coluna `location`.

```
df.location.unique()
```

Dentre as entradas da variável `location` existem países, continentes, distritos (não-países) e até `World`, que refere-se aos dados do mundo inteiro. Neste estudo serão considerados os países, continentes e os territórios isolados que, apesar de não serem considerados países, são relevantes à análise. Além dos dados globais, é claro.

Mas note que há uma série de entradas que não se encaixam nos quesitos mencionados. Vamos removê-las.

```
df_clean = df.drop(df.loc[df['location'].isin(['European Union', 'High income', 'International', 'Low income', 'Lower middle income', 'Upper middle income'])].index)
```

Agora que já removemos algumas colunas e entradas que não são relevantes ao estudo, vamos verificar a porcentagem de valores ausentes para cada coluna. 

```
(df_clean.isnull().sum() / df_clean.shape[0]).sort_values(ascending = False)
```

Até então não podemos assumir nenhum dos dados nulos como inconsistentes, visto que os mesmos podem representar, por exemplo, o não registro de novas mortes ou casos naquela data e localidade específica. Vamos prosseguir com a análise dos dados.

### Análise Global

Para iniciar iremos fazer uma breve análise do panorama da COVID-19 de maneira global. Vamos começar avaliando o total de casos e óbitos registrados.

```
# criando DataFrame específico para os dados globais
df_global = df_clean[(df_clean['location'] == 'World')]
df_global.set_index('date', inplace=True);
```

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

df_global['total_cases'].plot(kind='line', color='gray', ax=ax1)
ax1.set_ylabel('Total de casos registrados')
ax1.set_xlabel('Data')
ax1.set_title('Casos Registrados - Global')
ax1.ticklabel_format(style='plain', axis='y')

df_global['total_deaths'].plot(kind='line', color='gray', ax=ax2)
ax2.set_ylabel('Total de mortes registradas')
ax2.set_xlabel('Data')
ax2.set_title('Mortes Registradas - Global')
ax2.ticklabel_format(style='plain', axis='y')

df_global['new_cases_smoothed'].plot(kind='line', color='gray', ax=ax3)
ax3.set_ylabel('Quantidade de casos registrados')
ax3.set_xlabel('Data')
ax3.set_title('Novos Casos Registrados - Global')
ax3.ticklabel_format(style='plain', axis='y')

df_global['new_deaths_smoothed'].plot(kind='line', color='gray', ax=ax4)
ax4.set_ylabel('Quantidade de mortes registradas')
ax4.set_xlabel('Data')
ax4.set_title('Novas Mortes Registradas - Global');

plt.tight_layout()
```

Pelo gráfico do total de casos podemos verificar que, apesar de já ocorrerem diversas flexibilizações em relação às medidas preventivas da COVID-19, houve um acréscimo significativo nos casos em 2022. Isso pode ser confirmado pelo gráfico de novos casos registrados. Note que há um grande picos de novos casos registrados em janeiro-fevereiro de 2022.

Por sua vez, o gráfico de total de mortes por COVID-19 já apresenta uma leve curvatura para baixo, desde o final de 2021 até hoje. Essa situação fica evidente ao avaliarmos o gráfico de novas mortes registradas, apesar do pico de fevereiro-março de 2022, há uma tendência clara de diminuição no registro de novos óbitos.

Vamos evidenciar essa correlação entre casos e óbitos:

```
correlacao_casos_mortes = df_global[['total_cases', 'total_deaths']].corr()
display(correlacao_casos_mortes)
```

Certo, conseguimos visualizar de uma forma bem abrangente a situação dos casos e óbitos por COVID-19. Mas e a situação das vacinas? Vamos verificar.

```
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))

df_global['people_fully_vaccinated'].plot(kind='line', color='gray', ax=ax1)
ax1.set_ylabel('Pessoas com pelo menos duas doses da vacina')
ax1.set_xlabel('Data')
ax1.set_title('Pessoas Vacinadas (2 doses) - Global')
ax1.ticklabel_format(style='plain', axis='y')

df_global['new_people_vaccinated_smoothed'].plot(kind='line', color='gray', ax=ax2)
ax2.set_ylabel('')
ax2.set_xlabel('Data')
ax2.set_title('Novas Pessoas Vacinadas (2 doses) - Global')
ax2.ticklabel_format(style='plain', axis='y')

plt.tight_layout()
```

Pelos gráficos de vacinação pode-se identificar que, apesar da curva do total de pessoas vacinadas, está ocorrendo uma redução brusca no número de novas pessoas vacinadas com a segunda dose (ou dose única). Isso nos demonstra uma situação alarmante, visto que, em 31/03/2022, apenas 58,03% da população, vide abaixo, está vacinada com a segunda dose ou dose única.

```
# percentual de pessoas com a segunda dose ou dose única no mundo
(df_global.loc['2022-03-31'].people_fully_vaccinated / df_global.loc['2022-03-31'].population)*100
```

Todos sabemos da importância da vacina e do seu impacto positivo na melhora do quadro pandêmico da COVID-19. Mas para comprovarmos isso, vamos verificar a correlação entre o total de pessoas vacinadas com a segunda dose ou dose única com o registro de novos óbitos.

```
correlacao_mortes_pessoasvacinadas = df_global[['people_fully_vaccinated', 'new_deaths_smoothed']].corr()
display(correlacao_mortes_pessoasvacinadas)
```

Note que há uma correlação significativamente forte e inversa entre as duas variáveis. Significa que conforme o número de pessoas vacinadas aumenta, o número de novos óbitos diminui.

Conseguimos entender como está a situação dos casos, óbitos e vacinas de maneira global. Agora vamos fracionar a análise por continente.

### Análise por Continente

Nessa segunda etapa, vamos analisar os dados de 2022 para cada continente. Vamos começar verificando quais os continentes com o maior número de casos e de mortes para a data mais recente do nosso `dataset` (31/03/2022).

```
df_clean[df_clean['iso_code'].str.len() == 8].loc[df_clean.date == '2022-03-31'].sort_values(by="new_cases", ascending=False)
```

Note que o país *Kosovo* está, diferentemente dos outros países, com a sigla *OWID_* na frente do seu `iso_code` e, por isso, está aparecendo junto aos dados globais e continentais. Outro ponto é que iremos alterar o `iso_code` dos dados globais, para facilitar o filtro dos dados continentais. Vamos alterar isso e prosseguir com a análise.

```
# alterando o iso_code de Kosovo para 'KOS'
df_clean.loc[df_clean.iso_code == 'OWID_KOS', 'iso_code'] = 'KOS'

# alterando o iso_code dos dados globais (World) [OWID_WRL] para 'OWID_WORLD'
df_clean.loc[df_clean.iso_code == 'OWID_WRL', 'iso_code'] = 'OWID_WORLD'
```

```
df_clean[df_clean['iso_code'].str.len() == 8].loc[df_clean.date == '2022-03-22'].sort_values(by="new_cases", ascending=False)
```

```
df_clean[df_clean['iso_code'].str.len() == 8].loc[df_clean.date == '2022-03-22'].sort_values(by="new_deaths", ascending=False)
```

Para facilitar a visualização e comparação dos dados, vamos plotar gráficos para os registros de novos casos e mortes na data mais recente do `dataset` no momento desta análise, 31 de março de 2022.

```
# gráfico de barras para registros de casos e mortes por continente, para o dia 31/03/2022
df_continentes_31032022 = pd.DataFrame(df_clean[df_clean['iso_code'].str.len() == 8].loc[df_clean.date == '2022-03-31'])
casosrecentes_continentes = df_continentes_31032022.sort_values(by="new_cases", ascending=False)
mortesrecentes_continentes = df_continentes_31032022.sort_values(by="new_deaths", ascending=False)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

ax1.bar('location', 'new_cases', data=casosrecentes_continentes, color=['blue', 'yellow', 'green', 'red', 'pink', 'black'], edgecolor='black')
ax1.set_ylabel('Qtd. novos casos')
ax1.set_xlabel('Continente')
ax1.set_title('Novos casos registrados por continente em 31/03/2022')
ax1.ticklabel_format(style='plain', axis='y')

ax2.bar('location', 'new_deaths', data=mortesrecentes_continentes, color=['blue', 'yellow', 'red', 'pink', 'green', 'black'], edgecolor='black')
ax2.set_ylabel('Qtd. novas mortes')
ax2.set_xlabel('Continente')
ax2.set_title('Novas mortes registradas por continente em 31/03/2022')

plt.tight_layout()
```

Ambos os gráficos já nos trazem informações interessantes, mas podem ser ardilosas. Não podemos avaliar com os números brutos de um dia específico a real situação de cada continente. Para irmos um pouco mais à frente em nossa análise, vamos gerar os gráficos dos registros de casos e mortes no ano de 2022 para cada continente.

```
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8), (ax9, ax10), (ax11, ax12)) = plt.subplots(nrows=6, ncols=2, figsize=(20, 24))

df_europa_2022['new_cases_smoothed'].plot(kind='line', color='blue', ax=ax1)
ax1.set_ylabel('Quantidade de casos registrados')
ax1.set_xlabel('Data')
ax1.set_title('Casos registrados em 2022 na Europa')
ax1.ticklabel_format(style='plain', axis='y')

df_europa_2022['new_deaths_smoothed'].plot(kind='line', color='blue', ax=ax2)
ax2.set_ylabel('Quantidade de mortes registradas')
ax2.set_xlabel('Data')
ax2.set_title('Mortes registradas em 2022 na Europa')

df_asia_2022['new_cases_smoothed'].plot(kind='line', color='#DAA520', ax=ax3)
ax3.set_ylabel('Quantidade de casos registrados')
ax3.set_xlabel('Data')
ax3.set_title('Casos registrados em 2022 na Ásia')

df_asia_2022['new_deaths_smoothed'].plot(kind='line', color='#DAA520', ax=ax4)
ax4.set_ylabel('Quantidade de mortes registradas')
ax4.set_xlabel('Data')
ax4.set_title('Mortes registradas em 2022 na Ásia')

df_oceania_2022['new_cases_smoothed'].plot(kind='line', color='green', ax=ax5)
ax5.set_ylabel('Quantidade de casos registrados')
ax5.set_xlabel('Data')
ax5.set_title('Casos registrados em 2022 na Oceania')

df_oceania_2022['new_deaths_smoothed'].plot(kind='line', color='green', ax=ax6)
ax6.set_ylabel('Quantidade de mortes registradas')
ax6.set_xlabel('Data')
ax6.set_title('Mortes registradas em 2022 na Oceania')

df_sa_2022['new_cases_smoothed'].plot(kind='line', color='#FF00FF', ax=ax7)
ax7.set_ylabel('Quantidade de casos registrados')
ax7.set_xlabel('Data')
ax7.set_title('Casos registrados em 2022 na América do Sul')

df_sa_2022['new_deaths_smoothed'].plot(kind='line', color='#FF00FF', ax=ax8)
ax8.set_ylabel('Quantidade de mortes registradas')
ax8.set_xlabel('Data')
ax8.set_title('Mortes registradas em 2022 na América do Sul')

df_na_2022['new_cases_smoothed'].plot(kind='line', color='red', ax=ax9)
ax9.set_ylabel('Quantidade de casos registrados')
ax9.set_xlabel('Data')
ax9.set_title('Casos registrados em 2022 na América do Norte')

df_na_2022['new_deaths_smoothed'].plot(kind='line', color='red', ax=ax10)
ax10.set_ylabel('Quantidade de mortes registradas')
ax10.set_xlabel('Data')
ax10.set_title('Mortes registradas em 2022 na América do Norte')

df_africa_2022['new_cases_smoothed'].plot(kind='line', color='black', ax=ax11)
ax11.set_ylabel('Quantidade de casos registrados')
ax11.set_xlabel('Data')
ax11.set_title('Casos registrados em 2022 na África')

df_africa_2022['new_deaths_smoothed'].plot(kind='line', color='black', ax=ax12)
ax12.set_ylabel('Quantidade de mortes registradas')
ax12.set_xlabel('Data')
ax12.set_title('Mortes registradas em 2022 na África')

plt.tight_layout()
```

```
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

df_europa_2022['new_cases_smoothed'].plot(kind='line', color='blue', label=df_europa_2022.location[0], ax=ax1)
df_europa_2022['new_deaths_smoothed'].plot(kind='line', color='blue', label=df_europa_2022.location[0], ax=ax2)
df_asia_2022['new_cases_smoothed'].plot(kind='line', color='#DAA520', label=df_asia_2022.location[0], ax=ax1)
df_asia_2022['new_deaths_smoothed'].plot(kind='line', color='#DAA520', label=df_asia_2022.location[0], ax=ax2)
df_oceania_2022['new_cases_smoothed'].plot(kind='line', color='green', label=df_oceania_2022.location[0], ax=ax1)
df_oceania_2022['new_deaths_smoothed'].plot(kind='line', color='green', label=df_oceania_2022.location[0], ax=ax2)
df_sa_2022['new_cases_smoothed'].plot(kind='line', color='#FF00FF', label=df_sa_2022.location[0], ax=ax1)
df_sa_2022['new_deaths_smoothed'].plot(kind='line', color='#FF00FF', label=df_sa_2022.location[0], ax=ax2)
df_na_2022['new_cases_smoothed'].plot(kind='line', color='red', label=df_na_2022.location[0], ax=ax1)
df_na_2022['new_deaths_smoothed'].plot(kind='line', color='red', label=df_na_2022.location[0], ax=ax2)
df_africa_2022['new_cases_smoothed'].plot(kind='line', color='black', label=df_africa_2022.location[0], ax=ax1)
df_africa_2022['new_deaths_smoothed'].plot(kind='line', color='black', label=df_africa_2022.location[0], ax=ax2)

ax1.set_ylabel('Quantidade de casos registrados')
ax1.set_xlabel('Data')
ax1.set_title('Casos registrados em 2022 por continente')
ax1.legend()
ax1.ticklabel_format(style='plain', axis='y')

ax2.set_ylabel('Quantidade de mortes registradas')
ax2.set_xlabel('Data')
ax2.set_title('Mortes registradas em 2022 por continente')
ax2.legend()

plt.tight_layout()
```

Conseguimos avaliar o comportamento da COVID-19, em 2022, em cada continente e compararmos os mesmos. Um outro ponto interessante para avaliarmos é o percentual de casos que resultam em óbitos, pois pode nos dar insights com relação à vacinação e a estrutura do local. Vamos verificar:

```
mortes_casos_europa = (df_europa_2022['new_deaths'].sum() / df_europa_2022['new_cases'].sum()) * 100
mortes_casos_asia = (df_asia_2022['new_deaths'].sum() / df_asia_2022['new_cases'].sum()) * 100
mortes_casos_oceania = (df_oceania_2022['new_deaths'].sum() / df_oceania_2022['new_cases'].sum()) * 100
mortes_casos_sa = (df_sa_2022['new_deaths'].sum() / df_sa_2022['new_cases'].sum()) * 100
mortes_casos_na = (df_na_2022['new_deaths'].sum() / df_na_2022['new_cases'].sum()) * 100
mortes_casos_africa = (df_africa_2022['new_deaths'].sum() / df_africa_2022['new_cases'].sum()) * 100

print('O percentual de casos que resultaram em óbitos na Europa, em 2022, é de: ', round(mortes_casos_europa, 2), '%')
print('O percentual de casos que resultaram em óbitos na Ásia, em 2022, é de: ', round(mortes_casos_asia, 2), '%')
print('O percentual de casos que resultaram em óbitos na Oceania, em 2022, é de: ', round(mortes_casos_oceania, 2), '%')
print('O percentual de casos que resultaram em óbitos na América do Sul, em 2022, é de: ', round(mortes_casos_sa, 2), '%')
print('O percentual de casos que resultaram em óbitos na América do Norte (e central), em 2022, é de: ', round(mortes_casos_na, 2), '%')
print('O percentual de casos que resultaram em óbitos na África, em 2022, é de: ', round(mortes_casos_africa, 2), '%')
```

Note que a África apresenta um percentual significativamente maior que os demais continentes. Vamos verificar como está o percentual de população vacinada de cada continente, visto que já vimos que existe uma correlação inversa entre a quantidade de pessoas vacinadas e o registro de novos óbitos.

```
populacao_vacina_europa = (df_europa_2022.loc['2022-03-31','people_fully_vaccinated'] / df_europa_2022.loc['2022-03-31','population']) * 100
populacao_vacina_asia = (df_asia_2022.loc['2022-03-31','people_fully_vaccinated'] / df_asia_2022.loc['2022-03-31','population']) * 100
populacao_vacina_oceania = (df_oceania_2022.loc['2022-03-31','people_fully_vaccinated'] / df_oceania_2022.loc['2022-03-31','population']) * 100
populacao_vacina_sa = (df_sa_2022.loc['2022-03-31','people_fully_vaccinated'] / df_sa_2022.loc['2022-03-31','population']) * 100
populacao_vacina_na = (df_na_2022.loc['2022-03-31','people_fully_vaccinated'] / df_na_2022.loc['2022-03-31','population']) * 100
populacao_vacina_africa = (df_africa_2022.loc['2022-03-31','people_fully_vaccinated'] / df_africa_2022.loc['2022-03-31','population']) * 100

print('O percentual de população vacinada na Europa, em 31/03/2022, é de: ', round(populacao_vacina_europa, 2), '%')
print('O percentual de população vacinada na Ásia, em 31/03/2022, é de: ', round(populacao_vacina_asia, 2), '%')
print('O percentual de população vacinada na Oceania, em 31/03/2022, é de: ', round(populacao_vacina_oceania, 2), '%')
print('O percentual de população vacinada na América do Sul, em 31/03/2022, é de: ', round(populacao_vacina_sa, 2), '%')
print('O percentual de população vacinada na América do Norte (e central), em 31/03/2022, é de: ', round(populacao_vacina_na, 2), '%')
print('O percentual de população vacinada na África, em 31/03/2022, é de: ', round(populacao_vacina_africa, 2), '%')
```

O percentual de população vacinada na África justifica o alto percentual de casos de COVID-19 que resultam em óbitos. Além, é claro, da estrutura hospitalar e dos recursos disponíveis em diversos países africanos.

Começaremos criando os DataFrames e as variáveis necessárias para a análise.

```
df_31122021 = pd.DataFrame(df_clean[(df_clean['date'] == '2021-12-31')])
df_31032022 = pd.DataFrame(df_clean[df_clean['date'] == '2022-03-31'])
df_31122021.set_index('iso_code', inplace=True)
df_31032022.set_index('iso_code', inplace=True)

# removendo alguns países que possuem dados em 31/12/2021 e não em 31/03/2022. Não são relevantes para a análise.
df_31122021.drop(['GGY', 'JEY', 'NRU', 'NIU', 'PRI', 'SXM'], inplace=True)

# criando a coluna do total de casos de 2022
df_31032022 = df_31032022.assign(total_cases_2022 = df_31032022['total_cases'] - df_31122021['total_cases'])
# criando a coluna do total de óbitos de 2022
df_31032022 = df_31032022.assign(total_deaths_2022 = df_31032022['total_deaths'] - df_31122021['total_deaths'])

# criando a coluna do percentual de casos em 2022 em relação à população do local
df_31032022 = df_31032022.assign(cases_2022_vs_population = (df_31032022['total_cases_2022'] / df_31032022['population'])*100)
# criando a coluna do percentual de óbitos em 2022 em relação aos casos de 2022
df_31032022 = df_31032022.assign(deaths_2022_vs_cases_2022 = (df_31032022['total_deaths_2022'] / df_31032022['total_cases_2022'])*100)
# criando a coluna do percentual de pessoas vacinadas em relação à população do local
df_31032022 = df_31032022.assign(vaccinated_population_vs_population = (df_31032022['people_fully_vaccinated'] / df_31032022['population'])*100)

df_paises_2022 = df_31032022[(df_31032022.index.str.len() == 3)]
```
